```python
#import libraries
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) #Adjusts the configuration of the plots we will create

#read in data

df = pd.read_csv(r'C:\Users\isaac\OneDrive\Documents\Portfolio Project\Python Movie Industry Project\movies.csv')
```

```python
#Lets look at the data

df.info()
```

```python
#Let's see if there's any missing data

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col,pct_missing))
```

```python
#drop missing data

df = df.dropna()
```

```python
#Check to see if it worked
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col,pct_missing))
```

```python
#Data types for our columns

df.dtypes
```

```python
# Fix data types for ease of calculations

df['budget'] = df['budget'].astype('int64')

df['gross'] = df['gross'].astype('int64')

```

```python
#Add correct year to fix mismatched year values

df['yearcorrect'] = df['released'].str.extract(pat = '([0-9]{4})').astype(int)

```

```python
#order by gross revenue

df = df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
# Drop any duplicates

df.drop_duplicates()
```

```python
#Done with cleaning, let's check correlation

#Budget may have high correlation with revenue
#Company may have high correlation with revenue

```

```python
#Build scatter plot with budget vs gross

plt.scatter(x=df['budget'], y=df['gross'])
plt.title('Budget vs Gross Earnings')
plt.xlabel('Gross Earnings')
plt.ylabel('Budget of Firm')
plt.show()
plt.close()
```

```python
#Plot Budget vs Gross regression plot

sns.regplot(x='budget', y='gross', data = df, scatter_kws={'color': 'red'}, line_kws={'color':'blue'})
```

```python
#Look deeper at correlation

df.corr(method='pearson') #Pearson,#Kendall,#Spearman
```

```python
#High correlation between budget and gross :)
```

```python
correlation_matrix = df.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Matrix')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
plt.close()
```

```python
#Company Correlation
df_numerized = df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name]= df_numerized[col_name].astype('category')
        df_numerized[col_name]= df_numerized[col_name].cat.codes

df_numerized
```

```python
correlation_matrix = df_numerized.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Matrix')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
plt.close()
```
